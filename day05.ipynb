```python
from IPython.display import display, Markdown
import itertools
import pandas
import numpy
import re
```

```python
file_name = "day05.txt"
text_file = open(file_name, "r")
blocks = text_file.read().split("\n\n")

header_lines = blocks[0].split('\n')
header_len = len(header_lines)
header_cols = header_lines[-1].split()

data_lines = blocks[1].split('\n')
data_len = len(data_lines)
```

```python
# display(Markdown('# Part 2 Answer is %d' % part2_answer))
```

```python
column_widths = [4] * len(header_cols)
df = pandas.read_fwf(file_name, nrows=header_len - 1, names=header_cols, header=None, widths=column_widths, keep_default_na=True)
```

```python
def move_crate(fr, to):
    popped = matrix[fr-1][0]
    matrix[fr-1].remove(matrix[fr-1][0])
    matrix[to-1] = [popped] + matrix[to-1]
    print(matrix)

def move_crates(fr, to, cn):
    popped = matrix[fr-1][:cn]
    matrix[fr-1] = matrix[fr-1][cn:]
    matrix[to-1] = popped + matrix[to-1]
    print(matrix)

def parse_rule(rule):
    m = re.match('move (?P<count>\d+) from (?P<from>\d+) to (?P<to>\d+)', rule)
    rule_count = int(m.group('count'))
    rule_from = int(m.group('from'))
    rule_to = int(m.group('to'))
    return { 'rule_count': rule_count, 'rule_from': rule_from, 'rule_to': rule_to}
```

```python
matrix = [list(df[column].dropna().str.replace(r'[\[\]]','', regex=True).values) for column in df]
rules = [parse_rule(line) for line in data_lines]
for rule in rules:
    print ('move %d from %d to %d' % (rule['rule_count'], rule['rule_from'], rule['rule_to']))
    for count in range(rule['rule_count']):
        move_crate(rule['rule_from'], rule['rule_to'])

top_crates = [item[0] for item in matrix]
part1_answer=''.join(top_crates)
display(Markdown('# Part 1 Answer is %s' % part1_answer))
```

```python
matrix = [list(df[column].dropna().str.replace(r'[\[\]]','', regex=True).values) for column in df]
rules = [parse_rule(line) for line in data_lines]
for rule in rules:
    print ('move %d from %d to %d' % (rule['rule_count'], rule['rule_from'], rule['rule_to']))
    move_crates(rule['rule_from'], rule['rule_to'], rule['rule_count'])

top_crates = [item[0] for item in matrix]
part2_answer=''.join(top_crates)
display(Markdown('# Part 2 Answer is %s' % part2_answer))
```
